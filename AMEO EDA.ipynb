## Univariate Analysis with Visuals 

A. Discrete Data (i.e. Categorical or Numerical Discrete Columns)

identify the numeric colums in the dataset

```python
def discrete_viz_analysis(data_col):
    
    fig, ax = plt.subplots(figsize=(15, 8), constrained_layout=True)
    fig.suptitle("Discrete Distribution Plot")
    
    ax.set_title("Count Plot")
    sns.barplot(x=data_col.index, y=data_col, ax=ax)
    
    plt.xticks(rotation=45)
    plt.show()
    

def continuous_viz_analysis(data_col):

    # Create a 2x2 grid of subplots
    fig, axes = plt.subplots(2, 2, figsize=(15, 10), constrained_layout=True)
    fig.suptitle(f"Continuous Distribution Plot for {data_col.name.title()}")
    
    # Plot a histogram of the data
    axes[0, 0].set_title("Histogram Plot")
    sns.histplot(data_col, ax=axes[0, 0])
    axes[0, 0].set_xlabel("Value")
    axes[0, 0].set_ylabel("Frequency")
    
    # Plot a kernel density estimate of the data
    axes[0, 1].set_title("KDE Plot")
    sns.kdeplot(data_col, fill=True, ax=axes[0, 1])
    axes[0, 1].set_xlabel("Value")
    axes[0, 1].set_ylabel("Density")
    
    # Plot a box plot of the data
    axes[1, 0].set_title("Box Plot")
    sns.boxplot(data_col, ax=axes[1, 0])
    axes[1, 0].set_xlabel("Value")
    
    # Plot a quantile-quantile plot of the data against a normal distribution
    axes[1, 1].set_title("QQ Plot - Normality Check")
    stats.probplot(data_col, dist=stats.norm, plot=axes[1, 1])
    axes[1, 1].set_xlabel("Theoretical Quantiles")
    axes[1, 1].set_ylabel("Ordered Values")
    axes[1, 1].grid()

    # Show the figure
    plt.show()
```

```python
for i in numeric_data.columns:
    continuous_viz_analysis(numeric_data[i])
    
```

## Outlier

```python
Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3 - Q1

mask = ~((numeric_data < (Q1 - 1.5 * IQR)) | (numeric_data > (Q3 + 1.5 * IQR))).any(axis=1)

# Apply the mask to get a DataFrame without outliers
df_cleaned = data[mask]


df_cleaned = df_cleaned.drop(columns=['mechanicalengg', 'electricalengg',
                                      'telecomengg','civilengg', 'id', 'collegeid', 
                                      'collegecityid', 'unnamed_0'])
```
